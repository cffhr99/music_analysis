# For Spotify dataset

## Create a list of Mood and put them into each data randomly as mockup

```python
Mood = ['Dark','Relaxing','Energetic','Happy','Aggressive']
from random import seed
from random import randint
import numpy as np
seed(1)
Mood_mockup = ["NA"]*spotify_mockup_df.shape[0]
list_value = np.arange(100)
for i in list(range(100)):
    list_value[i] = randint(0,4)
    Mood_mockup[i] = Mood[list_value[i]]
Mood_mockup_df = pd.DataFrame(Mood_mockup,columns=['Mood'])
Mood_Index_mockup_df = pd.DataFrame(list_value,columns=['Mood_Index'])
```

```python
spotify_mockup_df = spotify_mockup_df.merge(Mood_mockup_df,right_index=True,left_index=True)
spotify_mockup_df = spotify_mockup_df.merge(Mood_Index_mockup_df,right_index=True,left_index=True)
spotify_mockup_df.head()
```

## Mockup Model for Machine Learning

Variable 'Mood' has the real name of each mood and 'Mood_Index' is the numerical categories of moods.

```python
# Remove some non-numerical variables
spotify_mockup_df_cleaned = spotify_mockup_df.drop(labels=['Unnamed: 0','name','album','artist','release_date'],axis=1)
spotify_mockup_df_cleaned.head()
```

```python
x_columns = spotify_mockup_df_cleaned.drop(labels=['Mood','Mood_Index'],axis=1).columns
```

There are two options for machine learning model: Logistic Regression Moel and Neural Network Model

### Logistic Regression Model:
The basic logistic regression model only deals with binary classification. However, our dataset has 5 different types of mood. One idea is to transform each mood into boelen variable (0=no,1=yes) like one-hot encoded features. Then we can fit five logistic regression models for each mood to obtain a list possible mood for each song. The second idea is to use multinomial logistic regression which can deal with mutlti-classification situation.

#### Basic Logistic Regression Model:

```python
spotify_mockup_df_BLR = spotify_mockup_df_cleaned.drop(labels=['Mood_Index'],axis=1)
spotify_mockup_df_BLR.head(1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
spotify_mockup_df_cat = spotify_mockup_df_BLR.dtypes[spotify_mockup_df_BLR.dtypes == 'object'].index.tolist()
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)
# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(spotify_mockup_df_BLR[spotify_mockup_df_cat]))
# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names_out(spotify_mockup_df_cat)
encode_df.head()
```

```python
spotify_mockup_df_BLR = spotify_mockup_df_BLR.merge(encode_df,left_index=True,right_index=True)
spotify_mockup_df_BLR = spotify_mockup_df_BLR.drop(labels=spotify_mockup_df_cat,axis=1)
spotify_mockup_df_BLR.head()
```

Using Mood_Aggressive as exmaple

```python
# Split our preprocessed data into our features and target arrays
y = spotify_mockup_df_BLR['Mood_Aggressive'].values
X = spotify_mockup_df_BLR[x_columns].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the logistic regression model
log_classifier = LogisticRegression(solver="lbfgs",max_iter=200)

# Train the model
log_classifier.fit(X_train,y_train)

# Evaluate the model
y_pred = log_classifier.predict(X_test)
print(f" Logistic regression model accuracy: {accuracy_score(y_test,y_pred):.3f}")
```

#### Mutinomial Logistic Regression

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
spotify_mockup_df_cleaned["Mood_Index"] = spotify_mockup_df_cleaned["Mood_Index"].astype(float)
y = spotify_mockup_df_cleaned["Mood_Index"].values
X = spotify_mockup_df_cleaned[x_columns].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# define the multinomial logistic regression model
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model.fit(X_train_scaled,y_train)
# report the model performance
y_pred = model.predict(X_test_scaled)
print(f" Logistic regression model accuracy: {accuracy_score(y_test,y_pred):.3f}")
```

#### Neural Network Model

```python
import tensorflow as tf
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 10
hidden_nodes_layer2 = 6

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=5, activation="softmax"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
from tensorflow import keras 
y_int = y_train.astype(int)
y_train_cat = keras.utils.to_categorical(y_int)
y_test_cat = keras.utils.to_categorical(y_test.astype(int))
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train_cat,epochs=50)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test_cat,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
